```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
driver=webdriver.Chrome(r'C:/Users/DELL/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
url='https://www.flipkart.com/'
driver.get(url)
    
```

```python
search_sunglasses=driver.find_element_by_class_name("_3704LK")
search_sunglasses
```

```python
search_sunglasses.send_keys("Sunglasses")
```

```python
search_btn=driver.find_element_by_xpath('/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_btn
```

```python
search_btn.click()
```

```python
title_tags=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
len(title_tags)
title_tags
```

```python
brand=[]
for i in title_tags:
    brand.append(i.text)
brand
```

```python
product_description=driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
len(product_description)
product_description
```

```python
prod_desc=[]
for i in product_description:
    prod_desc.append(i.text)    
prod_desc
```

```python
discount_prc=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
len(discount_prc)
discount_prc
```

```python
Discount_Price=[]
for i in discount_prc:
    Discount_Price.append(i.text)
    
Discount_Price
```

```python
desc=driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]')
len(desc)
desc
```

```python
Discount=[]
for i in desc:
    Discount.append(i.text)
    
Discount
```

```python
print(len(brand),len(prod_desc),len(Discount_Price),len(Discount))
```

```python
sunglasses=pd.DataFrame()
sunglasses['Brand']=brand
sunglasses['Product Description']=prod_desc
sunglasses['Discount Price']=Discount_Price
sunglasses['Discount']=Discount
sunglasses
```

```python
sunglasses.to_csv("Sunglasses listed on Flipkart.csv")
```

```python
driver.close()
```
